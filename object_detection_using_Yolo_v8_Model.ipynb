```python
# ========== DRAW GROUND TRUTH POLYGONS ==========
with open(label_path, 'r') as f:
    for line in f.readlines():
        parts = list(map(float, line.strip().split()))
        cls_id = int(parts[0])
        coords = parts[1:]

        # Convert normalized polygon coords to pixel coords
        polygon = []
        for i in range(0, len(coords), 2):
            x = int(coords[i] * w)
            y = int(coords[i+1] * h)
            polygon.append((x, y))

        # Draw polygon in green
        pts = np.array(polygon, dtype=np.int32)
        cv2.polylines(image, [pts], isClosed=True, color=(0, 255, 0), thickness=2)

        # Put class label
        x, y = polygon[0]
        cv2.putText(image, f"GT: mobject", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,0), 1)


# ========== DRAW PREDICTED BOUNDING BOXES ==========
for box in results[0].boxes:
    x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()
    conf = float(box.conf[0])
    label = f"Pred: mobject ({conf:.2f})"
    cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (255, 0, 0), 2)
    cv2.putText(image, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 1)

# ========== SHOW FINAL IMAGE ==========
plt.figure(figsize=(10, 8))
plt.imshow(image)
plt.axis('off')
plt.title("Green = Ground Truth (Polygon), Blue = YOLOv8 Prediction (Box)")
plt.show()
```

```python
import os
import glob
import cv2
import numpy as np
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, classification_report

# ========== CONFIGURATION ==========
model_path = '/content/drive/MyDrive/object-detection/yolov8_train/weights/best.pt'
image_dir = '/content/drive/MyDrive/object detection.v3i.yolov8/test/images'
label_dir = '/content/drive/MyDrive/object detection.v3i.yolov8/test/labels'
class_names = ['mobject']
iou_threshold = 0.5  # for TP/FP matching

model = YOLO(model_path)

# ==== IoU Calculation ====
def compute_iou(boxA, boxB):
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])
    interArea = max(0, xB - xA) * max(0, yB - yA)
    boxAArea = (boxA[2] - boxA[0]) * (boxA[3] - boxA[1])
    boxBArea = (boxB[2] - boxB[0]) * (boxB[3] - boxB[1])
    iou = interArea / (boxAArea + boxBArea - interArea + 1e-6)
    return iou

# ==== Evaluation Variables ====
tp, fp, fn = 0, 0, 0
y_true = []
y_pred = []

image_paths = glob.glob(os.path.join(image_dir, '*.jpg'))
print(f"Evaluating on {len(image_paths)} test images...\n")

for image_path in image_paths:
    filename = os.path.basename(image_path)
    label_path = os.path.join(label_dir, filename.replace('.jpg', '.txt'))

    img = cv2.imread(image_path)
    h, w = img.shape[:2]

    # ==== Load Ground Truth Polygons ====
    gt_boxes = []
    if os.path.exists(label_path):
        with open(label_path, 'r') as f:
            for line in f.readlines():
                values = list(map(float, line.strip().split()))
                cls_id = int(values[0])
                coords = values[1:]

                # Convert polygon to bbox
                polygon = np.array([(coords[i]*w, coords[i+1]*h) for i in range(0, len(coords), 2)])
                x_min, y_min = polygon.min(axis=0)
                x_max, y_max = polygon.max(axis=0)
                gt_boxes.append([x_min, y_min, x_max, y_max])

    # ==== Predict ====
    results = model.predict(image_path, conf=0.3, iou=0.5, verbose=False)
    pred_boxes = results[0].boxes.xyxy.cpu().numpy().tolist()

    matched_gt = set()
    for pred_box in pred_boxes:
        matched = False
        for i, gt_box in enumerate(gt_boxes):
            iou = compute_iou(pred_box, gt_box)
            if iou >= iou_threshold and i not in matched_gt:
                tp += 1
                matched_gt.add(i)
                matched = True
                break
        if not matched:
            fp += 1
    fn += len(gt_boxes) - len(matched_gt)

# ==== METRICS ====
print("==== EVALUATION RESULTS ====")
print(f"True Positives (TP): {tp}")
print(f"False Positives (FP): {fp}")
print(f"False Negatives (FN): {fn}\n")

precision = tp / (tp + fp + 1e-6)
recall = tp / (tp + fn + 1e-6)
f1 = 2 * (precision * recall) / (precision + recall + 1e-6)

print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1 Score:  {f1:.4f}")

# ==== Confusion Matrix ====
y_true = [1]*tp + [1]*fn + [0]*fp
y_pred = [1]*tp + [0]*fn + [1]*fp

print("\nConfusion Matrix:")
print(confusion_matrix(y_true, y_pred, labels=[1,0]))

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=['background', 'mobject']))
```
